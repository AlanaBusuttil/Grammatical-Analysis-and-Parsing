```python
import nltk
import nltk.grammar
import nltk.parse
import nltk.data
import copy

from collections import defaultdict
from copy import deepcopy

from nltk import CFG
from nltk.grammar import Nonterminal
from nltk.parse import EarleyChartParser
from nltk import tree, treetransforms
```

```python
#loading my grammar
g=nltk.data.load("./grammar.cfg.txt")

#instance of parser with grammar 
grammar = CFG.fromstring(g)
grammar.productions()

ec=nltk.parse.EarleyChartParser(grammar)


#reading each sentence and parsing it
sentences="./sentences.txt"
avg=[]
with open(sentences) as s:
    while s:
        getSentences = s.readline()
        sentence=str(getSentences)

        sentence=sentence.replace('.', ' .')
        sentence=sentence.replace('?', ' ?')
        sentence=sentence.replace('%', ' %')

        sentence1 = sentence.split()
        count=0
        for t in ec.parse(sentence1):
        #printing out a simple braceted structure and number of parses
                print(t)
                count=count+1
                print(count, "parses")
                avg=count
                
       # t = tree.Tree.fromstring(sentence, remove_empty_top_bracketing=True)
                collapsedTree = deepcopy(t)
                treetransforms.collapse_unary(collapsedTree)
                cnfTree = deepcopy(collapsedTree)
                treetransforms.chomsky_normal_form(cnfTree)

average=0
div=range(0,len(avg))
for x in div:
    average=average+x
avgParse=average/div
print("the average number of parses is:", avgParse)    
```
